```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create plot and add label
plt.plot(x_axis, y_axis, label="Boston")
#label axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
#set limits
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, "gD-", linewidth=2, label="Boston")
#g for green, D for diamond, - for line
ax.legend()
ax.set_xlabel('Date')
ax.set_ylabel('Fares($)')
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0, 45)
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
#formatted plot
plt.bar(x_axis, y_axis, color="green", label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#horizontal
# Create the plot.
plt.barh(x_axis, y_axis, color = "magenta", label="Boston")
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# object oriented method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='k', label='Boston')
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
ax.invert_yaxis()
ax.legend()
ax.set_ylabel('Date')
ax.set_xlabel('Fares($)')
ax.set_title("PyBer Fare by Month")
```

```python
# Scatter
#option 1: plt.plot(x_axis, y_axis, 'o')
plt.scatter(y_axis, x_axis, color='red', label="Chicago")
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
plt.xlim(0, 50)
```

```python
# Scatter bubble
#y_axis_larger = []
#for data in y_axis:
    #y_axis_larger.append(data*3)
    

plt.scatter(y_axis, x_axis, color='red', s=[i * 3 for i in y_axis], label="Chicago")
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
plt.xlim(0, 50)
```

```python
#object oriented
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color="skyblue", s=[i * 5 for i in y_axis], alpha=0.8, edgecolors="black", linewidth=2, label="Boston")
ax.invert_yaxis()
ax.set_xlim(0, 50)
ax.legend()
ax.set_ylabel('Date')
ax.set_xlabel('Fares($)')
ax.set_title("PyBer Fare by Month")
```

```python
# increase size (needs to be before plt.pie command
plt.subplots(figsize=(8, 8))
# explode values same sumber as number of slices values from 0 to 1, how far from center
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#Create colot list
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
#to add percent labels use the autopct
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#add shadow, add start angle, add colors
ax.pie(y_axis,explode=explode_values, colors=colors, shadow=True, startangle=-95, labels=x_axis,autopct='%.1f%%')
#invert order
ax.invert_yaxis()
plt.show()
```
